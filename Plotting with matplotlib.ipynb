```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(500)
y = np.random.randn(500)
```

```python
plt.scatter(x, y)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title('gauss')
plt.show()
```

```python
# make a figure with 2 axes to be plotted within it (i.e. 2 rows...the 1st param to subplot)
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1) # Could also create manually sized Axes object
ax.scatter(x, y)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_title('Gaussian Data')

ax2 = fig.add_subplot(2, 1, 2) # Could also create manually sized Axes object
ax2.scatter(x, y)
ax2.set_xlim(-1, 1)
ax2.set_ylim(-1, 1)
ax2.set_title('Zoomed')

plt.show()
```

```python
import math
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
t = np.linspace(0, 10, 300)
x = np.sin(2 * math.pi * 0.1 * t**2)
ax.step(t, x)
```

```python
x = np.random.randn(500)
y = np.random.randn(500)


from matplotlib.lines import Line2D
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line = Line2D(x,y)
ax.add_artist(line)
```

```python
# this resets the aspect
ax.autoscale_view()
ax.set_aspect(1.0, 'datalim')
line.set_color('green')
line.set_linestyle('dashed')
ax.add_line(line)
plt.draw()
display(fig)
```
